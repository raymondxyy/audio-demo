```python
# Usage: modify the parameters in this block and click Kernel -> Restart & Run All
PATH = '../samples/welcome16k.wav'  # path to your audio sample
QUANTIZE_NBIT = 8  # [1, 16] is a reasonable range
```

```python
import IPython.display as ipd

ipd.Audio(PATH, rate=16000)  # Orignal audio
```

```python
from util import quantize
import soundfile as sf

assert (QUANTIZE_NBIT >= 1) and (int(QUANTIZE_NBIT) == QUANTIZE_NBIT), "Unsupported input!"
sig, sr = sf.read(PATH)
sig_nbit = quantize(sig, int(QUANTIZE_NBIT))
```

```python
ipd.Audio(sig_nbit, rate=sr)  # Quantized audio. WARNING: it may get LOUD!
```

```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt
font = {'family' : 'serif',
        'weight' : 'bold',
        'size'   : 14}
plt.rc('font', **font)

fig, ax = plt.subplots(figsize=(16, 8))
ax.plot(np.arange(len(sig))/sr, sig, 'b')
ax.plot(np.arange(len(sig))/sr, sig_nbit, 'r')
plt.legend(['Original', f'{QUANTIZE_NBIT}-bit quantized'])
```
